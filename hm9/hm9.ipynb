```python
def download_image(url):
    """Download image from URL"""
    with request.urlopen(url) as resp:
        buffer = resp.read()
    stream = BytesIO(buffer)
    img = Image.open(stream)
    return img

def prepare_image(img, target_size):
    """Resize image to target size"""
    if img.mode != 'RGB':
        img = img.convert('RGB')
    img = img.resize(target_size, Image.NEAREST)
    return img
```

```python
import ssl
import urllib.request
from io import BytesIO
from PIL import Image

ssl_context = ssl.create_default_context()
ssl_context.check_hostname = False
ssl_context.verify_mode = ssl.CERT_NONE

def download_image(url):
    """Download image from URL"""
    with urllib.request.urlopen(url, context=ssl_context) as resp:
        buffer = resp.read()
    stream = BytesIO(buffer)
    img = Image.open(stream)
    return img
```

```python
test_url = "https://habrastorage.org/webt/yf/_d/ok/yf_dokzqy3vcritme8ggnzqlvwa.jpeg"
img = download_image(test_url)
```

```python
img_resized = prepare_image(img, target_size)
print(f"Resized to: {img_resized.size}")

# Convert to numpy array
x = np.array(img_resized, dtype='float32')
print(f"Array shape: {x.shape}")
print(f"Array dtype: {x.dtype}")
```

```python
mean = np.array([0.485, 0.456, 0.406])
std = np.array([0.229, 0.224, 0.225])

# Scale to [0, 1]
x = x / 255.0
x = (x - mean) / std
x = x.transpose(2, 0, 1)
x = np.expand_dims(x, axis=0)
```

```python
first_pixel_r = x[0, 0, 0, 0]
```

```python
options = [-10.73, -1.073, 1.073, 10.73]
closest = min(options, key=lambda opt: abs(opt - first_pixel_r))
```

```python
print(f" Q3 ANSWER: {closest}")
```
